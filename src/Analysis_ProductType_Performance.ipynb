```python
import pandas as pd
import matplotlib.pyplot as plt

#Load csv file into data frame
df = pd.read_csv(r'C:\Users\ngocd\source\Github\supply-chain-optimization\data\processed\merged_data.csv')

#Analyze sales performance by product types
product_sales = df.groupby(['product_id','name'])['pieces_sold'].sum().reset_index() #group data by product_id, name, then sum no.of pieces sold, then reset dataframe to turn group keys into column

#Print sales performance by product
print('Sales performance of different products')
print(product_sales)

#Plot sales performances
plt.figure(figsize=(20,8))
plt.bar(product_sales['name'], product_sales['pieces_sold']) #create a bar chart with name as the x-axis and pieces_sold as y-axis
plt.title('Sales performance of different products')
plt.xlabel('Product name')
plt.ylabel('Number of pieces sold')
plt.xticks(rotation=45, ha='right') #rotate the label of x-axis to 45 degree and align it to the right
plt.grid(True)
plt.show()
```
