```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\ngocd\source\Github\Coding-projects_July2024\data\processed\merged_data.csv')

#Calculate the delay in shipping
df['shipping_delay'] = df['actual_shipping_time'] - df['expected_shipping_time']

#Group by weight and calculate the mean delay and delay risks
performance_by_weight = df.groupby('weight_kg').agg({'shipping_delay':'mean','delay_risk':'mean'}).reset_index()

#Print the performance analysis by weight
print(performance_by_weight)

#Plot the analysis
plt.figure(figsize=(10,6))
plt.plot(performance_by_weight['weight_kg'],
performance_by_weight['shipping_delay'], marker='o', label='Mean Shipping Delay')
plt.plot(performance_by_weight['weight_kg'], performance_by_weight['delay_risk'], marker='x', label='Mean Delay Risk')
plt.xlabel('Weight(kg)')
plt.ylabel('Days/Risk')
plt.title('Transportation Analysis')
plt.legend()
plt.grid(True)
plt.show()
```
